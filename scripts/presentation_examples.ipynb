An example of the effects of similarity thresholding, implemented in python.

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Illustration of the heaviside function
def heaviside(x, threshold):
    """
    Heaviside step function
    """
    if x > threshold:
        return 1
    else: 
        return 0

def heaviside_scaling(x, threshold):
    """
    Heaviside step function with scaling
    """
    if x > threshold:
        return x
    else: 
        return 0

# Create a range of x values
x = np.linspace(-1, 1, 1000)
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Plot the first function
axs[0].plot(x, [heaviside(i, 0.6) for i in x])
axs[0].set_xlabel('S')
axs[0].set_ylabel(r"$f(S) = \theta(S - 0.6)$")
axs[0].legend()
axs[0].set_title('Heaviside Step Function')

# Plot the second function
axs[1].plot(x, [heaviside_scaling(i, 0.6) for i in x])
axs[1].set_xlabel('S')
axs[1].set_ylabel(r'$f(S) = \theta(S - 0.6) S$')
axs[1].set_title('Heaviside Step Function with Scaling')

plt.tight_layout()
```
