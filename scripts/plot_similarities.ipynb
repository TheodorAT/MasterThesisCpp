```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
forplan_original="../similarity_logs/forplan_PDLP"
forplan_improved="../similarity_logs/forplan"
bnl2_original="../similarity_logs/bnl2_PDLP"
bnl2_deteriorated="../similarity_logs/bnl2"

files = [forplan_original, forplan_improved, bnl2_original, bnl2_deteriorated]

similarity_dict = {}
for file_instance in files:
    similarities = []
    similarities_rel_first = []
    with open(file_instance, "r") as file:
        lines = []
        for line in file:
            lines.append(line)
        similarities_str = lines[0].split(",")[1:-1]
        similarities_str_rel_first = lines[1].split(",")[:-1]
    for sim in similarities_str:
        similarities.append(float(sim.strip()))
    for sim in similarities_str_rel_first:
        similarities_rel_first.append(float(sim.strip()))
    key = file_instance.split("/")[-1]
    similarity_dict[key] = similarities
    print(key, len(similarities))
    rel_key = key + "_rel_first"
    similarity_dict[rel_key] = similarities_rel_first
    print(rel_key, len(similarities_rel_first))
```

```python
start = 0
end = 100

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
fig.suptitle('forplan PDLP (Steering vectors = good)')
ax1.plot(similarity_dict["forplan_PDLP"][start:end])
ax2.plot(similarity_dict["forplan_PDLP_rel_first"][start:end])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
fig.suptitle('forplan +Steering (Steering vectors = good)')
ax1.plot(similarity_dict["forplan"][start:end])
ax2.plot(similarity_dict["forplan_rel_first"][start:end])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
fig.suptitle('bnl2 PDLP (Steering vectors = bad)')
ax1.plot(similarity_dict["bnl2_PDLP"][start:end])
ax2.plot(similarity_dict["bnl2_PDLP_rel_first"][start:end])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
fig.suptitle('bnl2 +Steering (Steering vectors = bad)')
ax1.plot(similarity_dict["bnl2"][start:end])
ax2.plot(similarity_dict["bnl2_rel_first"][start:end])
```
